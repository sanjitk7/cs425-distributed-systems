## Part 2: Put, get, delete operation latency

```python
dict = {'':['1', '2', '3', '4', '5'],
        'put':[451, 342, 396, 339, 544],
        'get':[221, 361, 205, 338, 396],
        'delete': [97, 170, 132, 158, 146],
       }

df = pd.DataFrame(dict)
  
display(df)
```

```python
df.describe()
```

```python
plt.plot([1,1,1,1,1], df["put"], label="put")
plt.plot([2,2,2,2,2], df["get"], label="get")
plt.plot([3,3,3,3,3], df["delete"], label="delete")
plt.xlabel("Trial #")
plt.ylabel("Latency time (measured in milliseconds)")
plt.title("File size 25MB")
plt.legend(loc="upper right")
fig = plt.figure(figsize =(10, 7))
 
# Creating plot
bp = df.boxplot(figsize=(7,5), showmeans=True)
bp.set_title('File Size 25 MB')
bp.set_xlabel('Operation type')
bp.set_ylabel('Latency time (measured in milliseconds)')
```

```python
dict = {'':['1', '2', '3', '4', '5'],
        'put':[7.42, 9.75, 5.89, 8.86, 8.95],
        'get':[6.63, 4.7, 5.14, 4.3, 5.6],
        'delete': [1.54, 0.91, 1.27, 1.62, 1.79],
       }

df = pd.DataFrame(dict)
  
display(df)
```

```python
plt.plot([1,1,1,1,1], df["put"], label="put")
plt.plot([2,2,2,2,2], df["get"], label="get")
plt.plot([3,3,3,3,3], df["delete"], label="delete")
plt.xlabel("Trial #")
plt.ylabel("Latency time (measured in seconds)")
plt.title("File size 500 MB")
plt.legend(loc="upper right")
fig = plt.figure(figsize =(10, 7))
 
# Creating plot
bp = df.boxplot(figsize=(7,5), showmeans=True)
bp.set_title('File Size 500 MB')
bp.set_xlabel('Operation type')
bp.set_ylabel('Latency time (measured in seconds)')
```

## Part 3: num-versions

```python
dict2 = {
'1': [209, 192, 231, 218, 236],
'2': [328, 387, 362, 356, 310],
'3': [408, 496, 366, 487, 424],
'4': [705, 693, 602, 569, 536],
'5': [930, 826, 954, 691, 926]}
df2 = pd.DataFrame(dict2)

plt.plot([1,1,1,1,1], [209, 192, 241, 258, 236], label="1")
plt.plot([2,2,2,2,2], [328, 277, 241, 258, 310], label="2")
plt.plot([3,3,3,3,3], [328, 277, 241, 258, 310], label="3")
plt.plot([4,4,4,4,4], [328, 277, 241, 258, 310], label="4")
plt.plot([5,5,5,5,5], [328, 277, 241, 258, 310], label="5")
plt.xlabel("num-version")
plt.ylabel("Latency time (measured in milliseconds)")
plt.title("get-versions latency")
plt.legend(loc="upper right")
fig = plt.figure(figsize =(10, 7))
 
# Creating plot
bp = df2.boxplot(figsize=(7,5), showmeans=True)
bp.set_title('get-versions latency')
bp.set_xlabel('num-version')
bp.set_ylabel('Latency time (measured in milliseconds)')
```
